### Part 3: Calculate the average price of an order.

Hint: Examine the data to see if the `quantity` column is relevant to this calculation.

Hint: Think carefully about the simplest way to do this!

```python
# checking on quantity in second column col[1] - if all quantities are equal to 1, it does not matter
quantity = []
for row in split_data:
    quantity.append(float(row[1])) # turning strings into floats so they can be used for arithmetic operations
print(quantity)
len(quantity)
```

Data shows that quantities are anywhere between 1 and 10. Quantities do matter for the calculation of average prices per order.

```python
# extract the amount paid per order, i.e. per line in the dataset
paid = []
for row in split_data:
    paid.append(float(row[4])) # turning strings into floats so they can be used for arithmetic operations
print(paid)
len(paid)
```

```python
# the two prior lists have the same length
# we can divide each $-price in list 'paid' by the corresponding item in list 'quantity'
average_price = []
for i in range(0,len(paid)):
    price = paid[i]/quantity[i]
    average_price.append(price)
print(average_price)
```

```python
# overall average
total_average = sum(average_price) / len(average_price)
total_average
```

The overall average price per order is about **US$ 7.08** (taking the quantities ordered into account)

### Part 4: Create a list (or set) named `unique_sodas` containing all of unique sodas and soft drinks that Chipotle sells.

Note: Just look for `'Canned Soda'` and `'Canned Soft Drink'`, and ignore other drinks like `'Izze'`.

```python
# create a list and check in the respective position in each row, i.e. third column row[2] for the two key words
# if the drink is already part of the list, it does not have to be passed into the list again and can be skipped
unique_sodas = []
for row in split_data:
    if row[2] == 'Canned Soda' or row[2] == 'Canned Soft Drink':
        if row[3] in unique_sodas: # skipping drink if already in list
            None
        else:
            unique_sodas.append(row[3])
    else:
        None
print(unique_sodas)
```

```python
# for kicks created a new dictionary that uses the brand names as keys and counts the occurence as values
# ignoring the quantity of orders here; only counting frequency
soda_count = {}
for brand in unique_sodas:
    soda_count[brand] = []

for i in range(0,len(unique_sodas)):
    count = 0
    for row in split_data:
        if row[3] == unique_sodas[i]:
            count += 1
    soda_count[unique_sodas[i]].append(count)
print (soda_count)
```

```python
# sorting the soda count from lowest value to highest value
import operator
sorted_soda = sorted(soda_count.items(), key=operator.itemgetter(1)) # itemgetter(1) sorts on values; itemgetter(0) on keys
print(sorted_soda)
```

Apparently **Coke** and **Diet Coke** are the most ordered canned beverages at Chipotle with **123** and **134** orders respectively.

```python
# let's do the same, but take the order quantity into consideration
soda_count_all = {}
for brand in unique_sodas:
    soda_count_all[brand] = []

for i in range(0,len(unique_sodas)):
    count = 0
    for row in split_data:
        if row[3] == unique_sodas[i]:
            count += 1*float(row[1])
    soda_count_all[unique_sodas[i]].append(count)
print (soda_count_all)
```

```python
# sorting soda count all from lowest value to highest value
sorted_soda_all = sorted(soda_count_all.items(), key=operator.itemgetter(1))
print (sorted_soda_all)
```

Biggest difference for Sprite, Coke and Diet Coke - apparently customers like to order multiple of these together (as a group). The volumes increase for:
    \tSprite: 77 to 89
    \tCoke: 123 to 143
    \tDiet Coke: 134 to 159

### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


```python
toppings = []

for row in split_data:
    if "Burrito" in row[2]:
        toppings.append(row[3])
toppings
```

```python
# from our list of toppings, the easiest way is to calculate the ',' as some toppings consist of multiple words
# in every order_description, the toppings are separated by a ',' irrespective of the number of words used to describe topping

toppings_count = []

for row in toppings:
    count = 1 # need to start the counter at 1, because there is +1 item vs the sum of commas in each row
    for item in row:
        if item == ',':
            count += 1
        else:
            pass
    toppings_count.append(count)
print(toppings_count)
```

```python
toppings_total_average = sum(toppings_count) / len (toppings_count)
toppings_total_average
```

On average, a burrito comes with **5.4 toppings** per order (ignoring quantity of burritos ordered, because that would not change the outcome dramatically)

### Part 6: Create a dictionary. Let the keys represent chip orders and the values represent the total number of orders.

Expected output: `{'Chips and Roasted Chili-Corn Salsa': 18, ... }`

Note: Please take the `quantity` column into account!

Optional: Learn how to use `.defaultdict()` to simplify your code.

```python
# replicating the *fun project* from earlier that was counting sodas by brand with chips

unique_chips = []
for row in split_data:
    if "Chips" in row[2]:
        if row[2] in unique_chips:
            None
        else:
            unique_chips.append(row[2])
print (unique_chips)
print (len(unique_chips))

chips_count_all = {}
for combos in unique_chips: #let's refer to different chips + other as combos
    chips_count_all[combos] = []

for i in range(0,len(unique_chips)):
    count = 0
    for row in split_data:
        if row[2] == unique_chips[i]:
            count += 1*float(row[1]) # multiplying count with quantity in that row to get the total number of orders
    chips_count_all[unique_chips[i]].append(count)

print (chips_count_all)
```

```python
# sorting chips count all from lowest value to highest value
sorted_chips_all = sorted(chips_count_all.items(), key=operator.itemgetter(1))
print (sorted_chips_all)
```

The most popular order for chips -- by a very long shot -- is **Chips with Guacamole**: **506** vs. **just Chips** in second place with **230** orders

```python
# now using .defaultdict() because I did not use that in the example above

from collections import defaultdict

chips_list = [] #store the entire list of chips here, i.e. no need to filter out duplicates because we'll count on that basis
for row in split_data:
    if "Chips" in row[2]:
        chips_list.append(row[2])

chips_count = defaultdict(int) # default value is zero 0
for chips in chips_list:
    chips_count[chips] += 1 # increments by values 1 every time it finds the same order

chips_count # call up the list by name spits out list of Chips orders sorted by key [equivalent to itemgetter(0)]
```

```python
# to make it easier to see smallest vs. largest item asked for, it's better to sort. Note: if itemgetter(0), it sorts by key
sorted_chips = sorted(chips_count.items(),key=operator.itemgetter(1))
sorted_chips
```
